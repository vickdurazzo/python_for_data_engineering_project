## Bibliotecas

```python
import requests
import pandas as pd
import json
```

## Extraindo dados usando uma API
Usando a famosa API de Pokemons nós iremos extrair as informações dos pokemons cadastrados na base.

```python
url = 'https://www.canalti.com.br/api/pokemons.json'
r = requests.get(url)
data = r.text
```

### Salvando como DataFrame

```python
type(data)
```

```python
data_dict = json.loads(data)
data_dict
```

```python
dataframe = pd.DataFrame(data_dict['pokemon'])
dataframe.head()
```

## Carregando os Dados
Usando o DataFrame salve como um CSV chamada `info_pokemons.csv`

```python
dataframe.to_csv("info_pokemons.csv")
```
